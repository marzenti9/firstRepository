```python
import PIL
import pytesseract
from pdf2image import convert_from_path
import matplotlib.pyplot as plt
import os
import nltk
```

```python
pdf_path = r'C:\Users\arzen\Downloads\PublicWaterMassMailing.pdf'
```

```python
images = convert_from_path(pdf_path, poppler_path=r'C:\Program Files\poppler-24.08.0\Library\bin')
```

```python
plt.figure(figsize = (5,5))
plt.imshow(images[0])
plt.show()
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\arzen\AppData\Local\Programs\Tesseract-OCR\tesseract.exe'
```

```python
all_text = ""
```

```python
os.chdir(r'C:\Users\arzen\OneDrive\Desktop')

for i, image in enumerate(images):
    text = pytesseract.image_to_string(image)
    all_text += text +"\n"
    with open(f'output_page_{i+1}.txt','w', encoding ='utf-8') as f:
        f.write(text)
```
